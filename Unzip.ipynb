```python
def UNZIP(FILEN, ex_dire):
    for i in range(len(FILEN)):
        fantasy_zip = zipfile.ZipFile(FILEN[i])
        fantasy_zip.extractall(ex_dire)
    fantasy_zip.close()
```

```python
FILEN = os.listdir("C:/Users/ATOMS8/Data/")[1:-2]
ex_dire = "C:/Users/ATOMS8/Data/"
UNZIP(FILEN, ex_dire)
```
